```python
import numpy as np

from functions.visualization_functions import *
from functions.ellipsoid_projection import *
```

This is a function to project a triaxial ellipsoid of arbitraty orientation along an arbitary direction.

Let's start by creating an ellipsoid, giving it three axis (eigenvectors) and their cooresponding axis lengths (eigenvalues).

```python
# normalized eigenvectors
eigenvec_min = np.asarray([0.73, 0.30, 0.61])
eigenvec_mid = np.asarray([-0.67, 0.14, 0.73])
eigenvec_max = np.asarray([-0.14, 0.94, -0.31])

eigenvecs = np.asarray([eigenvec_min, eigenvec_mid, eigenvec_max])

# eigenvalues
eigenvalues = np.asarray([.5, .7, 1])

# make sur the vectors are orthogonal
print(round(np.dot(eigenvec_min, eigenvec_mid), 2), round(np.dot(eigenvec_mid, eigenvec_max), 2))
```

This code was written to run on halos from the Abacus Summit simulations: https://github.com/abacusorg/AbacusSummit<br>
So we will formatt first formatt the ellipse as documented here: https://abacussummit.readthedocs.io/en/latest/data-products.html#halo-statistics

```python
ellipse = format_ellipsoid(eigenvecs, eigenvalues)
```

Before calculating the parameters of the projected ellipsid, let's take a look at it from a few different angles.<br>
This code plots a grid around the ellipse's surface

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(121)
plot_ellipsoid_grid(ellipse, proj_direction='x', point_size=1)
plt.xlabel('Y'); plt.ylabel('Z')

ax = fig.add_subplot(122)
plot_ellipsoid_grid(ellipse, proj_direction='z', point_size=1)
plt.xlabel('X'); plt.ylabel('Y')
ax.set_aspect('equal')
```

Notive how the ellipsoid axis don't exactly line up with the projected shape - ie you cannot simply roate the eigenvectors to get the projected axis

Now we need to choose the angle we wish to view this ellipse from. First, let's simply project along the $\hat{x}$ axis.

```python
a_projected, b_projected, angle_projected = get_projected_shape_x(ellipse)
```

...and now let's compare this with our original shape

```python
plot_E_flat(a_projected, b_projected, angle_projected*180/np.pi, scale=2.85)
plot_ellipsoid_grid(ellipse, proj_direction='x', point_size=1)

plt.xlabel('Y'); plt.ylabel('Z')
ax.set_aspect('equal')
```
